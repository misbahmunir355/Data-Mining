```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/Iris - Iris.csv")
data.head()
```

```python
dependent_variable = 'Species'
independent_variables = data.columns[data.columns != dependent_variable]
print("Dependent Variable:", dependent_variable)
print("Independent Variables:", independent_variables)
```

```python
print("Number of examples and features:")
print(data.shape)
```

```python
for column in data.columns:
    if data[column].dtype == 'object':
        print(f"\nValue counts for {column}:")
        print(data[column].value_counts())
```

```python
print("\nCounts for each feature:")
print(data.dtypes.value_counts())
```

```python
print("\nAttributes with missing values:")
print(data.isnull().sum())
```

```python
missing_values = data.isnull().sum()
if missing_values.any():
    print("Attributes with missing values:")
    print(missing_values)
else:
    print("No missing values in the dataset.")
```

```python
summary_stats = data.describe(include='all')
```

```python
print("Mean for each attribute:")
print(summary_stats.loc['mean'])
```

```python
print("\nMedian for each attribute:")
print(data.median())
```

```python
print("\nMode for each attribute:")
print(data.mode().iloc[0]) # Mode may have multiple values, so selecting the first one
```

```python
print("\nRange for each attribute:")
print(summary_stats.loc['max'] - summary_stats.loc['min'])
```

```python
print("\nStandard Deviation for each attribute:")
print(summary_stats.loc['std'])
```

```python
sns.pairplot(data, hue="Species")
plt.title("Pair Plot")
plt.show()
```

```python
sns.scatterplot(x="SepalLengthCm", y= "PetalLengthCm", hue="Species", data=data)
plt.show()
```

```python
data.hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Heatmap Correlation")
plt.show()
```
